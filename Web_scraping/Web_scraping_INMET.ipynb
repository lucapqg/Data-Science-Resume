```python
!pip install requests_html
from requests_html import HTMLSession 
from requests_html import AsyncHTMLSession
from datetime import date, datetime
import pandas as pd
```

```python
session = HTMLSession() # Initialize the GET requests
url = 'https://portal.inmet.gov.br/dadoshistoricos'
r = session.get(url) # GET requests
```

```python
[i for i in r.html.find('a') if 'https://portal.inmet.gov.br/uploads/dadoshistoricos' in str(i)]
```

```python
links = [i for i in r.html.find('a') if 'https://portal.inmet.gov.br/uploads/dadoshistoricos' in str(i)]
[{i:l} for i,l in  enumerate([i for i in r.html.find('a') if 'https://portal.inmet.gov.br/uploads/dadoshistoricos' in str(i)])]
```

```python
ultimo_link = len(links)-1
url = list(links[ultimo_link].absolute_links)[0]
```

```python
from io import  BytesIO
import requests
import zipfile

a = BytesIO(requests.get(url).content)
```

```python
# EXTRAI ARQUIVOS
mzip = zipfile.ZipFile(a)
# mzip.extractall("/content/drive/MyDrive/Python/Dados CSV")
```

```python
# NOME DE ARQUIVOS
[i.filename for i in mzip.infolist()]
```

```python
FILE_NAME = [i.filename for i in mzip.infolist() if 'DF' in i.filename] # A001 - ESTAÇÃO DE BRASILIA
FILE_NAME
```

```python
df_clim = pd.read_csv(mzip.extract(FILE_NAME[0]),skiprows = range(0, 8), error_bad_lines=False, encoding='latin-1',sep=';')
df_clim = df_clim.loc[:,~df_clim.columns.str.match("Unnamed")]
```

```python
df_clim.head(4)
```

```python
df_clim.info()
```

```python
df_clim.set_index('Data',inplace=True)# Seta o index para a nova coluna de data
df_clim.drop(['Hora UTC'],axis=1,inplace=True)# Exclui as colunas de data e hora inutilizaveis
```

```python
df_clim = df_clim.apply(lambda x: x.apply(lambda y: str(y).replace(',','.'))) #trocar todas as virgulas por pontos
df_clim = df_clim.astype(float) 
```
